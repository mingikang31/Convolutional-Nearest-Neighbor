# Results Heatmap using Python

```python
import seaborn as sns 
import matplotlib.pyplot as plt

import pandas as pd 
import numpy as np 
```

```python
data = np.random.rand(10, 12) 
df = pd.DataFrame(data, columns=[f'Var{i}' for i in range(1, 13)], index=[f'Obs{i}' for i in range(1, 11)])

```

```python
# Create a heatmap
sns.heatmap(df, annot=True, cmap='viridis')

# Display the heatmap
plt.show()
```
