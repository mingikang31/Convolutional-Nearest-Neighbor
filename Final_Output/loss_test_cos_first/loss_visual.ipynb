```python
import pandas as pd
import re
import os

def parse_training_results(file_path):
    """Parse training results from text file into a pandas DataFrame"""
    
    # Read the file
    with open(file_path, 'r') as f:
        content = f.read()
    
    # Initialize lists to store data
    epochs = []
    times = []
    train_losses = []
    train_acc_top1 = []
    train_acc_top5 = []
    test_losses = []
    test_acc_top1 = []
    test_acc_top5 = []
    
    # Regular expression to match each epoch line
    pattern = r'\[Epoch (\d+)\] Time: ([\d.]+)s.*?Train.*?Loss: ([\d.]+).*?Top1: ([\d.]+)%.*?Top5: ([\d.]+)%.*?Test.*?Loss: ([\d.]+).*?Top1: ([\d.]+)%.*?Top5: ([\d.]+)%'
    
    # Find all matches
    matches = re.findall(pattern, content)
    
    for match in matches:
        epochs.append(int(match[0]))
        times.append(float(match[1]))
        train_losses.append(float(match[2]))
        train_acc_top1.append(float(match[3]))
        train_acc_top5.append(float(match[4]))
        test_losses.append(float(match[5]))
        test_acc_top1.append(float(match[6]))
        test_acc_top5.append(float(match[7]))
    
    # Create DataFrame
    df = pd.DataFrame({
        'epoch': epochs,
        'time': times,
        'train_loss': train_losses,
        'train_accuracy_top1': train_acc_top1,
        'train_accuracy_top5': train_acc_top5,
        'test_loss': test_losses,
        'test_accuracy_top1': test_acc_top1,
        'test_accuracy_top5': test_acc_top5
    })
    
    return df
```

```python
# Categories 
data_set = ["VGG11-CIFAR10", "VGG11-CIFAR100"]
learning_rate = ["lr_1e-3", "lr_5e-4", "lr_1e-4", "lr_5e-5", "lr_1e-5"]
model_type = ["BranchingConvNN_K9_col_col_br0000_s42", "BranchingConvNN_K9_col_col_br0500_s42", "BranchingConvNN_K9_col_col_br1000_s42"]

for dataset in data_set:
    for lr in learning_rate:
        results = {}
        for model in model_type:
            ex_path = os.path.join(root_dir, "Final_Output", "loss_test", dataset, lr, model, "train_eval_results.txt")
            save_dir = os.path.join(root_dir, "Final_Output", "loss_test", "csv", "individual")
            save_path = os.path.join(root_dir, "Final_Output", "loss_test", "csv", "individual", f"{dataset}_{lr}_{model}.csv")

            os.makedirs(save_dir, exist_ok=True)


            df_results = parse_training_results(ex_path)
            df_results.to_csv(save_path, index=False)

            if model == "BranchingConvNN_K9_col_col_br0000_s42":
                model = "Conv2d"
            if model == "BranchingConvNN_K9_col_col_br1000_s42":
                model = "ConvNN"
            if model == "BranchingConvNN_K9_col_col_br0500_s42": 
                model = "Branching"

            results[model + "_Train_Loss"] = df_results["train_loss"]
            results[model + "_Test_Loss"] = df_results["test_loss"]
            results[model + "_Train_Accuracy"] = df_results["train_accuracy_top1"]
            results[model + "_Test_Accuracy"] = df_results["test_accuracy_top1"]
            
        results["epoch"] = df_results['epoch']
        save_dir = os.path.join(root_dir, "Final_Output", "loss_test", "csv", "lr")
        save_path = os.path.join(root_dir, "Final_Output", "loss_test", "csv", "lr", f"{dataset}_{lr}.csv")
        os.makedirs(save_dir, exist_ok=True)
        df = pd.DataFrame(results)
        df = df.reindex(columns=['epoch'] + [col for col in df.columns if col != 'epoch'])
        df.to_csv(save_path, index = False)
        
```
